# Importando

```python
import pandas as pd
```

# Dataset de exemplo

```python
import numpy as np

np.random.seed(42)

df = pd.DataFrame({
    'idade': np.random.randint(18, 65, size=100),
    'salario': np.random.normal(3000, 750, size=100).round(2),
    'valor': np.random.randint(10, 100, size=100),
    'sexo': np.random.choice(['Masculino', 'Feminino'], size=100),
    'classe': np.random.choice(['A', 'B', 'C'], size=100)
})
```

# Gráfico de linhas

```python
df[['idade', 'salario']].plot(figsize=(8, 4), title='Idade e Salário')
```

# Gráfico de barras

```python
df_ordenado = df.sort_values('classe', ascending=True)
df_ordenado['classe'].value_counts().plot(kind='bar', title='Frequência por Classe')
```

# Gráfico de pizza

```python
df['sexo'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Distribuição por classes')
```

```python
df['classe'].value_counts().plot(kind='pie',autopct='%1.1f%%', title='Distribuição por classe')
```

# Histograma

```python
df['salario'].plot(kind='hist', bins=15, color='skyblue', edgecolor='black', title='Distribuição de Salários')
```

# Boxplop gerado de uma série

```python
df['valor'].plot(kind='box', title='Boxplot de Valores')
```

# Area


```python
df[['idade', 'valor']].plot(kind='area', stacked=False)
```
