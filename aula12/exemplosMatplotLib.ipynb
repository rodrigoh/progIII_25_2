# Marcadores
* 'o': Círculo
* '.': Ponto
* ',': Pixel
* 'x': X
* '+': Mais
* '*': Estrela
* 's': Quadrado
* 'd': Diamante
* '^': Triângulo para cima
* 'v': Triângulo para baixo
* <: Triângulo para a esquerda
* '>': Triângulo para a direita
* 'h': Hexágono 1
* 'H': Hexágono 2
* 'p': Pentágono

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
plt.plot(x, y, marker='o', markersize=8, markerfacecolor='yellow', markeredgecolor='red')
```

# Títulos e rótulos


```python
x = [1, 2, 3, 4]
y1 = [2, 4, 1, 5]
y2 = [1, 3, 2, 4]

plt.plot(x, y1, label='Linha 1',  marker='o', markersize=8, markerfacecolor='yellow', markeredgecolor='red')
plt.plot(x, y2, label='Linha 2',  marker='*', markersize=8, markerfacecolor='yellow', markeredgecolor='red')

#Caso não especifique um local, fica na melhor posição, o valor padrão que é 'best'
plt.legend(loc='upper left')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico com Legenda')
plt.show()
```

# Legenda
*  'upper right': Canto superior direito
*  'upper left': Canto superior esquerdo
*  'lower right': Canto inferior direito
*  'lower left': Canto inferior esquerdo
*  'upper center': Centro superior
*  'lower center': Centro inferior
*  'center left': Centro esquerdo
*  'center right': Centro direito
*  'center': Centro do gráfico
*  'best': O Matplotlib tenta encontrar automaticamente a posição com a menor sobreposição dos dados.

```python
# Dados para as duas linhas
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Criando o gráfico
plt.plot(x, y1, label='Seno(x)')  # Adicionamos um rótulo para a primeira linha
plt.plot(x, y2, label='Cosseno(x)') # Adicionamos um rótulo para a segunda linha

# Adicionando a legenda
plt.legend()  # A função plt.legend() exibe a legenda com base nos rótulos definidos

# Adicionando rótulos aos eixos e um título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Seno e Cosseno')

# Exibindo o gráfico
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Gerando dados aleatórios para dois grupos
np.random.seed(42)  # Para garantir reprodutibilidade
x_grupo1 = np.random.rand(50)
y_grupo1 = np.random.rand(50)
x_grupo2 = np.random.rand(50) + 1  # Deslocando o segundo grupo no eixo x
y_grupo2 = np.random.rand(50) + 1  # Deslocando o segundo grupo no eixo y

# Criando o scatter plot para o Grupo 1
plt.scatter(x_grupo1, y_grupo1, label='Grupo A', marker='o', color='blue')

# Criando o scatter plot para o Grupo 2
plt.scatter(x_grupo2, y_grupo2, label='Grupo B', marker='s', color='red')

# Adicionando a legenda
plt.legend(title='Grupos')  # Podemos adicionar um título para a legenda

# Adicionando rótulos aos eixos e um título para o gráfico
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.title('Scatter Plot de Dois Grupos')

# Adicionando uma grade para melhor visualização
plt.grid(True, linestyle='--', alpha=0.5)

# Exibindo o gráfico
plt.show()
```
