# Importando e configurando

## Opções de estilo
* 'darkgrid' (padrão): Fundo cinza escuro com linhas de grade brancas.
* 'white': Fundo branco sem linhas de grade.
* 'dark': Fundo cinza escuro sem linhas de grade.
* 'ticks': Fundo branco com pequenos "ticks" (marcas) nos eixos, mas sem linhas de grade completas.

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Estilo visual (anteriormente se usava set)
sns.set_theme(style='whitegrid')
```

## Criando o dataframe de exemplo

```python
import pandas as pd
import numpy as np

# Criando um DataFrame de exemplo
np.random.seed(42)
df = pd.DataFrame({
    'idade': np.random.randint(18, 65, size=100),
    'salario': np.random.normal(3000, 750, size=100).round(2),
    'sexo': np.random.choice(['Masculino', 'Feminino'], size=100),
    'classe': np.random.choice(['A', 'B', 'C'], size=100),
    'categoria': np.random.choice(['Produto X', 'Produto Y', 'Produto Z'], size=100),
    'valor': np.random.randint(10, 100, size=100)
})
```

```python
df
```

# Gráfico de barras

```python
sns.barplot(x='categoria', y='valor', data=df)
plt.title("Barplot com Médias e Erros")
```

# Contagem de categorias

```python
sns.countplot(x='classe', data=df)
plt.title("Frequência por Classe")
```

```python
sns.countplot(x='classe', hue='sexo', data=df)
plt.title("Frequência por Classe e Sexo")
```

# Gráfico de dispersão

### hue='sexo': Este é um argumento poderoso do Seaborn que permite adicionar uma terceira dimensão categórica ao seu gráfico de dispersão através da cor. A coluna especificada aqui ('sexo') contém diferentes categorias ('Masculino', 'Feminino'). O Seaborn irá automaticamente atribuir uma cor diferente a cada categoria única presente na coluna 'sexo'. Isso permite visualizar se há alguma diferença na relação entre idade e salário com base no sexo. Por exemplo, você pode observar se homens e mulheres tendem a se agrupar em diferentes áreas do gráfico

```python
sns.scatterplot(x='idade', y='salario', hue='sexo', data=df)
```

# Boxplot

```python
sns.boxplot(x='sexo', y='idade', data=df)
```

# Mapa de calor de correlação

* annot=True: Este argumento booleano controla se os valores da matriz serão exibidos dentro de cada célula do heatmap.

 * Se **annot=True**, o valor da correlação (ou o valor correspondente na matriz de dados que você está visualizando) será escrito no centro de cada célula do heatmap. Isso torna a leitura dos valores subjacentes muito mais fácil e direta.
 * Se **annot=False**, o heatmap mostraria apenas as cores, sem os valores numéricos dentro das células. Isso pode ser útil para focar nos padrões gerais, mas torna a leitura de valores específicos mais difícil.

* cmap='coolwarm'
    * Algumas opções coolwarm, RdBu, bwr, seismic, PiYG, BrBG, RdYlBu, RdYlGn e Spectral (mais detalhes, ver documetação ou slides)
  

```python
#Calculando a correlção das variáveis numéricas
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='Spectral')
plt.title("Mapa de Calor das Correlações")
```
