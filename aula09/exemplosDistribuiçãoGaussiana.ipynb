```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Definindo a mdia e o desvio padro
media = 0
desvio_padrao = 1
tamanho_amostra = 1000

# Gerando dados aleatrios com distribuio normal
dados = np.random.normal(media, desvio_padrao, tamanho_amostra)

# Criando o histograma
plt.hist(dados, bins=30, density=True, alpha=0.6, color='b', label='Histograma dos Dados')

# Gerando pontos para a funo de densidade de probabilidade (PDF) da normal
x = np.linspace(media - 3*desvio_padrao, media + 3*desvio_padrao, 100)
pdf = (1/(desvio_padrao * np.sqrt(2 * np.pi))) * np.exp(-(x - media)**2 / (2 * desvio_padrao**2))

# Plotando a PDF
plt.plot(x, pdf, 'r', linewidth=2, label='PDF da Normal')

# Adicionando rtulos e ttulo
plt.title('Distribuio Normal (Gaussiana)')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)

# Exibindo o grfico
plt.show()
```

### Imagine que estamos analisando a altura de um grande grupo de mulheres adultas brasileiras. Estudos mostram que a altura mdia () dessa populao  de aproximadamente 1,60 metros (160 cm) e o desvio padro ()  de cerca de 6 centmetros. Podemos usar a Distribuio Normal para estimar a probabilidade de encontrar uma mulher com uma determinada altura dentro desse grupo.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definindo os parmetros da distribuio normal
media = 160
desvio_padrao = 6

# Criando um array de valores de altura para o eixo x
altura_min = media - 3 * desvio_padrao  # Considera 3 desvios padro para cobrir a maior parte da curva
altura_max = media + 3 * desvio_padrao
x = np.linspace(altura_min, altura_max, 100) # Cria 100 pontos entre os limites

# Calculando a funo de densidade de probabilidade (PDF) para cada valor de x
pdf = norm.pdf(x, media, desvio_padrao)

# Definindo os limites para a rea que queremos destacar (154 cm e 166 cm)
limite_inferior = 154
limite_superior = 166

# Criando arrays booleanos para identificar os valores dentro dos limites
regiao_interesse = (x >= limite_inferior) & (x <= limite_superior)

# Plotando a curva da distribuio normal
plt.plot(x, pdf, 'b', label=f'Mdia = {media} cm, Desvio Padro = {desvio_padrao} cm')

# Preenchendo a rea sob a curva entre os limites
plt.fill_between(x, pdf, where=regiao_interesse, color='orange', alpha=0.5, label=f'P({limite_inferior}  Altura  {limite_superior})')

# Adicionando linhas verticais nos limites
plt.axvline(limite_inferior, color='r', linestyle='--', linewidth=0.8)
plt.axvline(limite_superior, color='r', linestyle='--', linewidth=0.8)

# Adicionando rtulos e ttulo ao grfico
plt.xlabel('Altura (cm)')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuio Normal das Alturas de Mulheres Brasileiras Adultas')
plt.legend()
plt.grid(True)
plt.show()

# Calculando a probabilidade numericamente usando a funo cdf (Cumulative Distribution Function)
probabilidade = norm.cdf(limite_superior, media, desvio_padrao) - norm.cdf(limite_inferior, media, desvio_padrao)
print(f'\nA probabilidade de uma mulher ter altura entre {limite_inferior} cm e {limite_superior} cm  de aproximadamente: {probabilidade:.4f}')
```

O grfico representa a Distribuio Normal das Alturas de Mulheres Brasileiras Adultas. Aqui esto os principais elementos e suas interpretaes:

**Curva em Azul (Distribuio Normal)**: Esta curva em forma de sino  a representao visual da distribuio normal.

O pico da curva est localizado na mdia da distribuio, que  de 160 cm.
A largura da curva  determinada pelo desvio padro, que  de 6 cm. Uma curva mais estreita indicaria menor variabilidade nas alturas (as alturas seriam mais agrupadas em torno da mdia), enquanto uma curva mais larga indicaria maior variabilidade.

**Eixo Horizontal (Altura (cm))**: Este eixo representa os possveis valores de altura em centmetros. Vemos que as alturas variam de aproximadamente 142 cm at 178 cm na extenso mostrada, cobrindo uma faixa de valores em torno da mdia.

**Eixo Vertical (Densidade de Probabilidade)**: Este eixo representa a densidade de probabilidade para cada valor de altura.  importante notar que a altura da curva em um ponto especfico no  a probabilidade daquela altura exata (j que para variveis contnuas a probabilidade de um valor exato  zero). Em vez disso, a rea sob a curva em um determinado intervalo de alturas representa a probabilidade de uma mulher ter uma altura dentro desse intervalo.

**rea Sombreada em Laranja (P(154  Altura  166))**: A rea sob a curva que est sombreada em laranja representa a probabilidade de uma mulher adulta brasileira ter uma altura entre 154 cm e 166 cm.

As linhas verticais tracejadas em vermelho marcam os limites desse intervalo (154 cm e 166 cm).

Como calculamos anteriormente, essa rea corresponde a aproximadamente 68,27% da rea total sob a curva. Isso significa que h uma probabilidade de cerca de 68,27% de encontrar uma mulher adulta brasileira com altura entre 154 cm e 166 cm, assumindo que a altura dessa populao segue essa distribuio normal com a mdia e o desvio padro especificados.
