```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados de exemplo: Horas de estudo (x) e Nota (y)
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # reshape para virar matriz (obrigatório pro scikit-learn)
y = np.array([50, 55, 65, 70, 75])

# Cria e treina o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(x, y)

# Coeficientes da reta
a = modelo.intercept_   # intercepto
b = modelo.coef_[0]     # inclinação

print(f"Equação da reta: y = {a:.2f} + {b:.2f}x")

# Fazendo previsão (ex: para 6 horas de estudo)
horas_estudo = 6
nota_prevista = modelo.predict([[horas_estudo]])
print(f"Nota prevista para {horas_estudo} horas de estudo: {nota_prevista[0]:.2f}")

# Gráfico
plt.scatter(x, y, color='blue', label='Dados reais')
plt.plot(x, modelo.predict(x), color='red', label='Reta da regressão')
plt.xlabel('Horas de estudo')
plt.ylabel('Nota')
plt.title('Regressão Linear Simples')
plt.legend()
plt.grid(True)
plt.show()
```
