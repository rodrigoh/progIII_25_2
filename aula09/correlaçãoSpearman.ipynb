```python
import numpy as np
from scipy.stats import spearmanr

# Suponha que você tenha duas listas (ou arrays NumPy) representando seus dados
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 4, 6])

# Calcula a correlação de Spearman e o p-valor
correlacao_spearman, p_valor = spearmanr(x, y)

print(f"Os dados da variável X são: {x}")
print(f"Os dados da variável Y são: {y}")
print(f"O coeficiente de correlação de Spearman é: {correlacao_spearman:.4f}")
print(f"O p-valor associado é: {p_valor:.4f}")
```
