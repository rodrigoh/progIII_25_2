```python
#importando a biblioteca
import pandas as pd
```

```python
#Lendo o arquivo csv
df = pd.read_csv('dados.csv',index_col=0)
#Print do dataset
df
```

```python
#Verificando os valores únicos da coluna
df['Gênero'].unique()
```

```python
#Ajustando a coluna gênero
df['Gênero'] = df['Gênero'].replace({'Fem':'F', 'Masc':'M'})
```

```python
df.isnull().sum()
```

```python
#Pesquisando quem tem o gênero nulo
df.loc[df['Gênero'].isna()]
```

```python
#Resulver os nulos
df.loc[df['Nome']=='Marcos Pinto', 'Gênero'] = 'M' 
df.loc[df['Nome']=='Leonardo Almeida', 'Gênero'] = 'M' 
df.loc[df['Nome']=='Larissa Ribeiro', 'Gênero'] = 'F' 
df.loc[df['Nome']=='Cíntia Dias', 'Gênero'] = 'F' 
df.loc[df['Nome']=='Larissa Ribeiro', 'Gênero'] = 'F' 
```

```python
#Apagando as linhas duplicadas
print(df.shape)
df = df.drop_duplicates()
print(df.shape)
```

```python
df.loc[df['Nome']=='Larissa Ribeiro']
```

```python
#Preenchendo as idades
df['Idade'] = df['Idade'].fillna(df['Idade'].median())
df
```

```python
#Preenchendo o gsto mensal
df['Gasto Mensal'] = df['Gasto Mensal'].fillna(df['Gasto Mensal'].mean())
df
```

```python
df.info()
```

```python
#Exportar para um csv
df.to_csv('dados_processados.csv')
```
